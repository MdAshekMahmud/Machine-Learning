```python
import numpy as np
import pandas as pd

from sklearn.impute import KNNImputer, SimpleImputer
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv(
    r"C:\Users\infor\OneDrive\Desktop\Machine Learning\DataSets\Titanic-Dataset.csv"
)
df = df[["Age", "Pclass", "Fare", "Survived"]]
```

```python
df.head()
```

```python
X = df.drop(columns=["Survived"])
y = df["Survived"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=2)
```

```python
X_train.head()
```

```python
knn = KNNImputer(n_neighbors=3, weights="distance")

X_train_transform = knn.fit_transform(X_train)
X_test_transformed = knn.transform(X_test)
```

```python
pd.DataFrame(X_train_transform, columns=X_train.columns)
```

```python
lr = LogisticRegression()

lr.fit(X_train_transform, y_train)

y_pred = lr.predict(X_test_transformed)

accuracy_score(y_test, y_pred)
```

```python
# Comparison with mean

si = SimpleImputer()

X_train_trf2 = si.fit_transform(X_train)
X_test_trf2 = si.transform(X_test)
```

```python
lr = LogisticRegression()

lr.fit(X_train_trf2, y_train)

y_pred2 = lr.predict(X_test_trf2)

accuracy_score(y_test, y_pred2)
```
